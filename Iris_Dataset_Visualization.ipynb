```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
print(iris.shape)
print(iris.columns.tolist())
print(iris.head())
print(iris.info())
print(iris.describe())
```

```python
# Scatter Plot
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal vs Petal Length')
plt.show()

# Histogram
iris.hist(figsize=(10,6), color='steelblue')
plt.suptitle('Feature Distributions')
plt.show()

# Box Plot
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```

```python
print("=== KEY INSIGHTS ===")
print(f"Total Samples: {iris.shape[0]}")
print(f"Total Features: {iris.shape[1]-1}")
print(f"\nAverage Sepal Length: {iris['sepal_length'].mean():.2f}")
print(f"Average Petal Length: {iris['petal_length'].mean():.2f}")
print(f"\nSpecies Count:\n{iris['species'].value_counts()}")
```
